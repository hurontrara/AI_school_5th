```python
def compare(v1, v2):
    if v1 < v2: #증가율계산 
        try: value = round((v2 - v1)/v1*100,2)    # v1 < v2 일 때 얼마나 증가했는지 %
        except: value = 100                       # v1는 0이였고, v2는 증가했을 때 (100%)
    elif v1 > v2: #하락률 계산
        try: value = -round((v1 - v2)/v1*100,2) # v1 > v2 일 때 얼마나 하락했는지 %
        except: value = -100                    # v2가 0으로 하락했을 때 (-100%)
    else:        # 값이 같으면 0% 증감
        value = 0
    return value

rate_change = []

for i in range(len(rate)):
    rate_change.append(compare(rate["전체"][i],rate["전체"][i+1]))
# one_rate = compare(rate["전체"][2],rate["전체"][3])
# one_rate
rate_change
```
