```python
# 파이썬으로 데이터 주무르기 (민형기 저) 책 참조

import urllib.request
import urllib.parse
from bs4 import BeautifulSoup

import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
from wordcloud import WordCloud

# 키워드로 Naver 뉴스, 블로그 검색하여 word cloud 이미지 만들기

def main():

    searchList = ['코로나']
    sKeyword = ''.join(searchList)

    titleList = []    
    titleList = searchKeyword(sKeyword)
    allInOne = ' '.join(titleList)
    # print(len(titleList))
    
    maskSKorea = np.array(Image.open('SouthKorea2.jpg'))
                          
    wc = WordCloud(font_path='c:/Windows/Fonts/malgun.ttf',
                width=522, height=565, 
                background_color='white', colormap='rainbow_r',
                mask=maskSKorea, max_words=1500)
    # colormap='Accent_r', 'Dark2', 'Red' 'gist_rainbow' 

    wc = wc.generate(allInOne)
    array = wc.to_array()
    # print('array', array) #pixel별 색을 array로 저장함
    # wordcloud = wordcloud.generate_from_frequencies(keywords)

    fig = plt.figure(figsize=(10,10))
    plt.imshow(array, interpolation='bilinear')
    plt.axis('off')
    plt.show()
    fig.savefig('SKoreaWCImage.png')


def searchKeyword(keyword):
    
    baseUrl = 'https://search.naver.com/search.naver?'
    # optUrl = 'date_from={dFrom}&date_to={dEnd}&date_option=3&'
    endUrl = 'where=news&sm=tab_jum&query='
    plusUrl = f'{keyword}'
    startUrl = '&start={page}'
    # print('plusUrl', plusUrl)
        
    # 검색량이 많은 경우 pages 더 늘리기
    pages = [x for x in range(1, 102, 10)]
    # print(pages)

    # 네이버 news 기사 검색
    titleList = []
    for pp in pages:
        url = baseUrl + \
              endUrl + urllib.parse.quote_plus(plusUrl) + \
              startUrl.format(page=pp)
        print('url', url, '\n')

        html = urllib.request.urlopen(url).read()
        soup = BeautifulSoup(html, 'html.parser')
        # print(soup.prettify())

        for each in soup.find_all('a', \
                                  attrs={'class': '_sp_each_title'}):
            title = each.get_text()
            titleList.append(title)
            # print(title)

    # 네이버 블로그 posting 검색
    endUrl = 'where=post&sm=tab_jum&query='
    for pp in pages:
        url = baseUrl + \
              endUrl + urllib.parse.quote_plus(plusUrl) + \
              startUrl.format(page=pp)
        print('url', url, '\n')

        html = urllib.request.urlopen(url).read()
        soup = BeautifulSoup(html, 'html.parser')
        # print(soup.prettify())

        for each in soup.find_all('a', attrs={'class': '_sp_each_title'}):
            title = each.get_text()
            titleList.append(title)
            # print(title)
    return titleList

   
# main 함수 로딩부
if __name__ == '__main__':
    main()
```
